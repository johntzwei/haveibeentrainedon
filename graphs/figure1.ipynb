```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statistics

sns.set()
# sns.set_context('paper', font_scale=0.73)
sns.set_context('paper', font_scale=1.3)
sns.set_style('whitegrid', {'font.family':'serif', 'font.serif':'Times New Roman'})
```

```python
# Experiment directories
tradeoff_exp_df = pd.read_csv('/home/ryan/haveibeentrainedon/unstealthy/unstealthy_tradeoff/tradeoff_graph_data.csv')
repetition_wtmrklength_exp_df = pd.read_csv('/home/ryan/haveibeentrainedon/unstealthy/unstealthy_scaling/repetition_wtmrklength_graph_data.csv')
raretoken_exp_df = pd.read_csv('/home/ryan/haveibeentrainedon/unstealthy/unstealthy_raretoken/raretoken_graph_data.csv')
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)  # 1 row, 2 columns, 1st subplot
sns.lineplot(data=repetition_wtmrklength_exp_df, x="repetition", y="value", hue="exp_name", errorbar=("ci", 95), marker='o')

plt.subplot(1, 3, 2)  # 1 row, 2 columns, 1st subplot
ax = sns.lineplot(data=tradeoff_exp_df, x='Repetitions Per Watermark', y='Z-Score', errorbar=("ci", 95), marker='o')
ax.set_xscale("log")

# Creating the second plot (Bar plot)
plt.subplot(1, 3, 3)  # 1 row, 2 columns, 2nd subplot
sns.lineplot(data=raretoken_exp_df, x="Rank", y="Z-Score", errorbar=("ci", 95), marker='o')

plt.tight_layout()
plt.show()
```
